```python
import pandas as pd
pd.set_option('display.max_rows', None)
ghgs = pd.read_csv('UNdata_ghgs.csv')
```

```python
ghgs.rename(columns={'Country or Area':'country', 'Year':'id', 'Value':'value'}, inplace=True)
ghgs.head()
```

```python
ghgs = pd.crosstab(index=ghgs['country'], columns=ghgs['id'], values=ghgs['value'], aggfunc='sum')
ghgs.loc['Total'] = ghgs.sum()
ghgs = ghgs.reset_index()
ghgs.head()
```

```python
ghgs.to_csv('greenhouse_gases_1990_2019.csv', index=False, header=str)
```
