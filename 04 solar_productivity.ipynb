# Surplus vs European Deficit

```python
country_area = pd.DataFrame([('Germany', 357386),
                             ('EU & UK', 4475757),
                             ('Morocco', 710850),
                             ('North Africa', 6026150),
                             ('Brazil', 8515767)],
                            columns=('country','area'))
country_area
```

```python
energy_surplus = pd.DataFrame([('Morocco', 35777, 91066),
                             ('North Africa', 349305, 771998),
                             ('Brazil', 593576, 1090938)],
                            columns=('country','total_prod','projected_prod'))
energy_surplus
```

```python
energy_surplus['surplus'] = energy_surplus['projected_prod'] - energy_surplus['total_prod']
```

```python
energy_surplus
```

```python
eu_deficit = 1405875
```

```python
energy_surplus['percentage_eu_deficit'] = energy_surplus['surplus'] / eu_deficit
```
